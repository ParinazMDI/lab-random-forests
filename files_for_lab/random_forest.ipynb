# 2.  Feature Selections- Applying Variance Threshold

#Let's use this feature selection method on our main dataset. But before that, we need to scale our numeric data. If we try this method on the categorical data, we have to encode it first. However, encoding might reduce the varaice! Consequently,  there arises a possibility of considering the elimination of all categorical columns!!

```python
from sklearn.preprocessing import MinMaxScaler
transformer = MinMaxScaler().fit(num)
scaled_numerical = transformer.transform(num)
scaled_numerical = pd.DataFrame(scaled_numerical, columns=num.columns)

scaled_numerical
```

```python
from sklearn.feature_selection import VarianceThreshold

var_threshold = 0.04  # I consider this value for threshold to observe the impact on my model's performance
sel = VarianceThreshold(threshold=(var_threshold))

# 1) default is threshold = 0 :eliminate all columns that are identical for all rows
# 2) in practice we would scale the columns first, and then apply threshold, or apply
#    a different threshold for different columns

# This drops the columns that have a variance less than this threshold
sel.fit(scaled_numerical)
temp = sel.transform(scaled_numerical)
temp = pd.DataFrame(temp)
print(scaled_numerical.shape)
print(temp.shape)
```

```python
# we dropped (330-55) of our columns columns
pd.DataFrame(temp).head()
```

```python
# which columns are droped:
removed_columns = pd.DataFrame(data=(num.columns,sel.variances_,sel.get_support()), index=('column_name','variance','statement')).T

collist = list(removed_columns[removed_columns['statement'] == False]['column_name'])
print(collist)
len(collist)
```

```python
# now we would want to double check the correlation for each of these with the TARGET_B-we should drop, low variance_low cor
exc_list=[]
for col in collist:
    correlation=tar_df['TARGET_B'].corr(num[col])
    if correlation>0.02:
        exc_list.append(( correlation,col))
exc_list_df = pd.DataFrame(data=exc_list, columns=['correlation', 'column'])

        
exc_list_df
```

```python
# we can canclude that all collist has low variance and low cor
are_lists_equal = exc_list == collist
print(are_lists_equal)
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

features = cat.columns[:-1]  # Exclude the target column

significance_level = 0.05

for feature in features:
    contingency_table = pd.crosstab(cat[feature], tar_df['TARGET_B'])
    
    # Perform chi-squared test
    chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)
    
    # Print results for each feature
    print(f"\nChi-squared test for {feature}:")
    print(f"Chi2 Stat: {chi2_stat}")
    print(f"P-value: {p_value}")
    print(f"Degrees of Freedom: {dof}")
    
#     # Interpret results
    if p_value < significance_level:
        print("  There is a significant relationship between the feature and the target.")
    else:
        print("  There is no significant relationship between the feature and the target.")
```

```python
# going to drop TARGET_B column because I already selected the ones who donated

data = final_df.drop(columns=['TARGET_B'], axis=1)
```

```python
# X/y split
X = data.drop('TARGET_D',axis = 1)
y = data['TARGET_D']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0) 
X_train_num = X_train.select_dtypes(include=np.number)
X_train_cat = X_train.select_dtypes(include = object)


X_test_num = X_test.select_dtypes(include=np.number)
X_test_cat = X_test.select_dtypes(include = object)
```

```python
transformer = MinMaxScaler().fit(X_train_num) 
numericals_train_scaled = transformer.transform(X_train_num)
numericals_test_scaled = transformer.transform(X_test_num)
```

```python
encoder = OneHotEncoder(handle_unknown='error', drop='first').fit(X_train_cat)
categoricals_train_encoded = encoder.transform(X_train_cat).toarray()
categoricals_test_encoded = encoder.transform(X_test_cat).toarray()
```

```python
X_train = np.concatenate((numericals_train_scaled,categoricals_train_encoded),axis=1)
X_test = np.concatenate((numericals_test_scaled,categoricals_test_encoded),axis=1)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

lm = LinearRegression()
lm.fit(X_train,y_train)
predictions_test = lm.predict(X_test)
r2_score = r2_score(y_test, predictions_test) 
```

```python
train_r2 = lm.score(X_train, y_train)
test_r2 = lm.score(X_test, y_test)
print(f"Training R^2: {train_r2} -- Testing R^2: {test_r2}")
```

```python
# apply cross validation to the model
results = cross_validate(lm,X_train, y_train, cv = 5)
```
